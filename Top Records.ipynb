```python
#导入包(PypI)
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#加载数据
path="D:/学习教育/数据可视化/大作业/Top Records/"
os.chdir(path)

files = ["Top Records-PUBG.csv", "Top Records-CSGO.csv", "Top Records-Dota 2.csv"]
def load_data(file):
    df = pd.read_csv(file, encoding='utf-8')
    x = pd.to_datetime(df['DateTime'])
    y = df['Players']
    return x, y
data_dict = {}
for i, file_name in enumerate(files):
    x, y = load_data(file_name)
    var_name = f'x{i+1}', f'y{i+1}'
    data_dict[var_name] = (x, y)
data_dict
```

```python
x1, y1 = data_dict[('x1', 'y1')]
print(x1, y1)
```

```python
x2, y2 = data_dict[('x2', 'y2')]
print(x2, y2)
```

```python
x3, y3 = data_dict[('x3', 'y3')]
print(x3, y3)
```

```python
x = ['']*4242
y = [0]*4242
for i in range(0,4242):
    if i in range(0,1959):
        x[i] = x3[i]
    elif i in range(1959,4242):
        x[i] = x1[i-1959]
        y[i] = y1[i-1959]
x = [pd.NaT if val == '' else val for val in x]#将空字符串替换为NaN
#将x1、y1转换为Series类型
x1 = pd.Series(x)
y1 = pd.Series(y)
data_dict[('x1','y1')] = (x1, y1)
print(x1,y1)
```

```python
x = ['']*4242
y = [0]*4242
for i in range(0,4242):
    if i in range(0,69):
        x[i] = x3[i]
    elif i in range(69,4242):
        x[i] = x2[i-69]
        y[i] = y2[i-69]
x = [pd.NaT if val == '' else val for val in x]#将空字符串替换为NaN
#将x2、y2转换为Series类型
x2 = pd.Series(x)
y2 = pd.Series(y)
data_dict[('x2','y2')] = (x2, y2)
print(x2,y2)
```

```python
#绘制堆叠柱形图
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']#正常显示中文和特殊符号
plt.rcParams['axes.unicode_minus'] = False#正常显示负号
colors = ["#000094", "#8BACE3", "#95A7C7"]#自定义颜色列表
names=["PUBG: BATTLEGROUNDS","Counter-Strike: Global Offensive","Dota 2"]
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))#创建包含两个子图的图表对象
width = 1
axs[0].bar(x3, y3, width=width, color=colors[2])
axs[0].bar(x2, y2, bottom=y3, width=width, color=colors[1])
axs[0].bar(x1, y1, bottom=y2+y3, width=width, color=colors[0])
axs[0].set_xlabel('DateTime', size=15)
axs[0].set_ylabel('Players', size=15)
axs[0].set_title('堆叠柱形图', size=18)
axs[0].spines['right'].set_color('none')
axs[0].spines['top'].set_color('none')
axs[0].legend((names[2],names[1],names[0]), bbox_to_anchor=(0.6,1.0), frameon=False)
#绘制三折线图
axs[1].plot(x1, y1, color=colors[0])
axs[1].fill_between(x1, 0, y1, alpha=0.65, color=colors[0])
axs[1].plot(x2, y2, color=colors[1])
axs[1].fill_between(x2, 0, y2, alpha=0.65, color=colors[1])
axs[1].plot(x3, y3, color=colors[2])
axs[1].fill_between(x3, 0, y3, alpha=0.65, color=colors[2])
axs[1].set_xlabel('DateTime', size=15)
axs[1].set_ylabel('Players', size=15)
axs[1].set_title('三折线图', size=18)
axs[1].spines['right'].set_color('none')
axs[1].spines['top'].set_color('none')
axs[1].legend(names, bbox_to_anchor=(0.6,1.0), frameon=False)

plt.suptitle('Lifetime Players Count History', fontsize=26)
plt.savefig('堆叠柱形图和三折线图(缩放)-Top Records.png')
plt.show()
```
